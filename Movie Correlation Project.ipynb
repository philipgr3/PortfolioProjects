## Importing libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

## Reading the data

```python
df = pd.read_csv(r'C:\Users\Philip\analysis\Projects\Movies_Python\movies.csv')
```

## Checking for missing values

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
null_columns = df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```

```python
print(df[df.isnull().any(axis=1)][null_columns].head())
print(df[df.isnull().any(axis=1)][null_columns].tail())
```

## Dropping null values

```python
df = df.dropna()
df.head()
```

## Changing gross and budget to int

```python
df.dtypes
```

```python
df['gross'] = df['gross'].astype('int64')
df['budget'] = df['budget'].astype('int64')
df.head()
```

## Creating a new column for the correct year of release (some rows have different values when it comes to released date and year)

```python
df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
df.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

## Looking into correlations with scatter plots (budget vs gross)

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget')


plt.show()
```

## Using seaborn for regression plot

```python
sns.regplot(x='budget', y='gross', data = df, scatter_kws = {'color':'red'}, line_kws={'color':'black'})
```

## Looking more into correlation

```python
df.corr()
```

```python
# High correlation between budget and gross
```

## Seaborn heatmap

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

## Numerizing every object data value in order to see the correlation of all of our data values

```python
df_nummerized = df

for col_name in df_nummerized:
    if(df_nummerized[col_name].dtype == 'object'):
        df_nummerized[col_name] = df_nummerized[col_name].astype('category')
        df_nummerized[col_name] = df_nummerized[col_name].cat.codes
        
df_nummerized.head()     
```

```python
df.head()
```

```python
df_nummerized.corr()
```

```python
nummerized_correlation_matrix = df_nummerized.corr()
sns.heatmap(nummerized_correlation_matrix, annot = True)
plt.title('Correlation')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

## Taking a look into all correlations for each data value

```python
correlation_mat = df_nummerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

## Showing a list with the most correlated values

```python
high_corr = sorted_pairs[((sorted_pairs) > 0.5) & ((sorted_pairs) < 1)]
high_corr.sort_values(ascending = False)
```
